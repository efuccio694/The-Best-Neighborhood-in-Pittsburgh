## **Purpose:** Let's Set the Atmosphere, Shall we? 

**1.** I shall determine the neighborhoods with the highest concentration of stairways. This shall be calculated by using the City of Pittsburgh Steps dataset from the WRDC. 

**2.** I shall determine the neighborhoods with the stairways of longest lengths. 

**2.** I will then use these submetrics as an indication of the general fitness of the populations of Pittsburgh neighborhoods.

***

## **Data & Analysis:** Let's Dive in

```python
import pandas as pd
import numpy as np 
%matplotlib inline 
import matplotlib.pyplot as plt 
```

### **Our data initially appears as follows ...**

```python
staircases = pd.read_csv("CityOfPittsburghSteps.csv")
staircases.head(10)
```

### **Lets clean our data first ...**
    * redefine our dataframe to only include necessary columns
    * drop rows with missing data

```python
necessary = staircases[['neighborhood', 'length' , 'number_of_steps', 'latitude', 'longitude']]
necessary.head(10)
```

```python
cleaned_data = necessary.dropna() 
cleaned_data.head(10)
```

### **Lets create dictionaries to store the maximum values for our nbhds ...**
    * create dictionary for total number of steps for each nbhd
    * create dictionary for total length of steps for each nbhd

```python
d = dict()
for index, row in cleaned_data.iterrows():
    nbhd = row['neighborhood']
    num_steps = row['number_of_steps']
    if nbhd not in d:
        d[nbhd] = num_steps
    else:
        d[nbhd] = d[nbhd] + num_steps
for nbhd in d:
    print( str( nbhd) + ': ' + str(d[nbhd]) )
```

```python
j = dict()
for index, row in cleaned_data.iterrows():
    nbhd = row['neighborhood']
    length = row['length']
    if nbhd not in j:
        j[nbhd] = length
    else:
        j[nbhd] = j[nbhd] + length
for nbhd in j:
    print( str( nbhd) + ': ' + str(j[nbhd]) )
```

### **Now lets combine both of these dictionaries into one to create one data frame ...**

```python
length = pd.Series(j)
steps = pd.Series(d)
stairs_metrics = { "length": length, "steps": steps }
metrics = pd.DataFrame(stairs_metrics)
metrics
```

### **Let's now create two separate dataframes and plot their information ...**
    * one data frame for the 20 nbhds with largest step lengths
    * one data frame for the 20 nbhds with largest number of steps

```python
max_stairs = metrics.nlargest( 20, ['length'])
max_stairs = max_stairs.drop('steps',1)
max_stairs.plot(kind = 'line', title = "20 Highest Pitt NBHDS Step Length", legend = True,  rot =90)
```

```python
max_steps = metrics.nlargest( 20, ['steps'])
max_steps = max_steps.drop('length',1)
max_steps.plot(kind = 'line', title = "20 Highest Pitt NBHDS Steps Count", legend = True,  rot =90)
```
